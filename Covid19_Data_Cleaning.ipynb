```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Covid19.csv')
```

# Data Cleaning Process

```python
data.head(2)
# avg_stringency -> is the strict government policy Range from [0-100] = [low->High].
# avg_mobility -> movements from people -ve means low movement %, +ve means high movement %.  
```

```python
data.info()
```

```python
data.sample(5)
```

```python
data[data['confirmed'].isnull()]['country_name'].unique()
```

```python
data_cleaned = data[~data['country_name'].isin(['Antarctica', 'Bouvet Island', 'British Indian Ocean Territory',
       'Caribbean Netherlands', 'Christmas Island', 'Cocos Islands',
       'French Southern Territories', 'Heard Island and McDonald Islands', 'Netherlands Antilles', 'Norfolk Island',
       'South Georgia and the South Sandwich Islands',
       'Svalbard and Jan Mayen', 'United States Minor Outlying Islands',
       'Western Sahara'])]
```

```python
data_cleaned
```

```python
data_cleaned.info()
```

```python
data_cleaned = data_cleaned.fillna(0)
```

```python
data_cleaned.info()
```

```python
data_cleaned['confirmed'] = data_cleaned['confirmed'].astype('int32')
```

```python
data_cleaned['deaths'] = data_cleaned['deaths'].astype('int32')
```

```python
data_cleaned['tests'] = data_cleaned['tests'].astype('int32')
```

```python
data_cleaned['vaccinations'] = data_cleaned['vaccinations'].astype('int32')
```

```python
data_cleaned['elderly_pop'] = data_cleaned['elderly_pop'].astype('int32')
```

```python
data_cleaned['population'] = data_cleaned['population'].astype('int32')
```

```python
data_cleaned.info()
```

```python
data_cleaned.head(5)
```

```python
data_cleaned['infection_rate'] = (data_cleaned['confirmed']/data_cleaned['population'])*100
```

```python
data_cleaned['positivity_rate'] = (data['confirmed']/data['tests'])*100
data_cleaned.sample(5)
```

```python
data_cleaned['positivity_rate'] = data_cleaned['positivity_rate'].fillna(0)
```

```python
data_cleaned
```

```python
data_cleaned['date'] = pd.to_datetime(data_cleaned['year'].astype('str') + '-' + data_cleaned['month'].astype('str') + '-01')
```

```python
data_cleaned
```

```python
data_cleaned.to_csv("covid19_cleaned_data.csv", index=False)
```
