```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('covid19_cleaned_data.csv')
data.head(5)
```

# Top 10 countries with highest confirmed cases

```python
data_top_ten_confirm = data.groupby('country_name')['confirmed'].max().sort_values(ascending = False).head(10).reset_index()
data_top_ten_confirm
```

```python
sns.barplot(data = data_top_ten_confirm, x= 'country_name', y = 'confirmed')
plt.xticks(rotation = 75)
plt.show()
```

# Top 10 countries with highest deaths

```python
data_top_ten_death = data.groupby('country_name')['deaths'].max().sort_values(ascending = False).head(10).reset_index()
data_top_ten_death
```

```python
sns.barplot(data = data_top_ten_death, x= 'country_name', y = 'deaths')
plt.xticks(rotation = 75)
plt.show()
```

# Comparison of Top 10 countries with mobility


```python
data_mob = data[data['country_name'].isin(data_top_ten_confirm['country_name'].values)]
data_mob.head()
```

```python
sns.boxplot(data = data_mob, x = 'country_name', y = 'avg_mobility')
plt.xticks(rotation = 75)
plt.show()
```

# Comparison of Top 10 countries with stringency


```python
sns.boxplot(data = data_mob, x = 'country_name', y = 'avg_stringency')
plt.xticks(rotation = 75)
plt.show()
```
