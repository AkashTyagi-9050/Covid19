# average positivity rate (cases/tests) across all countries

```python
# average positivity formulae will be = no. of cases / no.of test for each country
# additional we will calculate positive percentage for each country formulae = ( 1 - average positivity )*100
```

```python
data_avg_positivity_rate = (data.groupby('country_name')['confirmed'].max() / data[data['tests'] != 0].groupby('country_name')['tests'].max()).reset_index(name = 'avg_positivity')
data_avg_positivity_rate
```

```python
data_avg_positivity_rate['avg_positivity_percentage'] = ((1- data_avg_positivity_rate['avg_positivity'])*100).round(2)
```

```python
data_avg_positivity_rate
```

# growth rate of confirmed cases between two countries

```python
# Compound Growth Rate -> CAGR=(First Value/Last Value​)**1/n − 1
```

```python
country_1 = data[data['country_name'] == 'India']
country_1
```

```python
country_2 = data[data['country_name'] == 'United States of America']
country_2
```

```python
country_1_CAGR = (((country_1['confirmed'].max()/country_1['confirmed'].min())**(1/24) -1 )*100)
country_1_CAGR
```

```python
country_2_CAGR = (((country_2['confirmed'].max()/country_2['confirmed'].min())**(1/24) -1 )*100)
country_2_CAGR
```
